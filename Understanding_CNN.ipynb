```python
import tensorflow as tf
from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = '/content/drive/MyDrive/Deep Learning/practical_Deep_learning/Chapter_2/Molly_006-2829x1886-2726x1886.jpg'
```

```python
img=image.load_img(image_path,target_size=(224,224))
```

```python
plt.imshow(img)
plt.show()
```

```python
image_array=image.img_to_array(img)
```

```python
image_array.shape
```

```python
image_array
```

```python
image_batch = np.expand_dims(image_array, axis=0)
```

```python
image_batch.shape
```

```python
model=tf.keras.applications.ResNet50(weights='imagenet')
```

```python
img_preprocessed = preprocess_input(image_batch)
```

```python
prediction = model.predict(img_preprocessed)
```

```python
print(decode_predictions(prediction, top=3)[0])
```

```python
def predict (image_path):
  img=image.load_img(image_path,target_size=(224,224))
  image_array=image.img_to_array(img)
  image_batch = np.expand_dims(image_array, axis=0)
  img_preprocessed = preprocess_input(image_batch)
  prediction = model.predict(img_preprocessed)
  return decode_predictions(prediction, top=3)[0]
```

```python
predict(image_path)
```
